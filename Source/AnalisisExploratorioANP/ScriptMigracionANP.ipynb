```python
import pandas as pd
```

```python
df=pd.read_excel('Estadísticas ANPs RMBA.xlsx')
```

```python
df.head(46)
```

```python
df= df[['Área Natural Protegida', 'Superficie (ha.)', 'Ecorregión',
       'Municipio', 'Localidad', 'Administración', 'Año de creación',
       'Categoría Internacional', 'Plan de Manejo', 'Guardaparques',
       'Infraestructura', 'Atención a visitantes']]
```

```python
toNewName ={
    'Área Natural Protegida':'Nombre', 'Superficie (ha.)':'Superficie', 'Ecorregión':'Ecoregion', 'Administración':'Administracion', 'Año de creación':'AnioCreacion',
       'Categoría Internacional':'CategoriaInternacional', 'Plan de Manejo':'PlanManejo', 'Atención a visitantes':'AtencionVisitantes'
}
df = df.rename(columns=toNewName)
```

# Depurando datos nulos

```python
df['Guardaparques'].fillna(value='No', inplace=True)
```

```python
df['Infraestructura'].fillna(value='No', inplace=True)
```

```python
df['AtencionVisitantes'].fillna(value='No', inplace=True)
```

```python
df['PlanManejo'].fillna(value='No', inplace=True)
```

```python
df['Superficie'].fillna(value=0, inplace=True)
```

```python
df['Ecoregion'].replace('Delta e Islas del Paraná','Delta e islas del Paraná',inplace=True)
```

```python
df['Ecoregion'].replace('Pampa y Delta e Islas del Paraná','Pampa y Delta e islas del Paraná',inplace=True)
```

```python
df[0:46].to_json(path_or_buf='anp.json', orient='records')
```
